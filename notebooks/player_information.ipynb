```python
remove_nonFIFA_teams = ["Comoros", "Czechoslovakia", "Faroe Islands", "French Guiana", "German DR", "Guadeloupe", 
                           "Kosovo", "Macau", "Martinique", "Papua New Guinea", "Saarland", "Tahiti", 
                           "Vietnam Republic", "Yemen DPR", "Yugoslavia"]

df = df[~df["team"].isin(remove_nonFIFA_teams)]
df = df[~df["home_team"].isin(remove_nonFIFA_teams) & ~df["away_team"].isin(remove_nonFIFA_teams)]
```

```python
df.tail(20) 
```

```python
new_data_eng = pd.DataFrame({
    'date': pd.to_datetime('2026-07-18'), 
    'home_team': 'France',                
    'away_team': 'England',               
    'team': 'England',                    
    'scorer': ['Declan Rice', 'Ezri Konsa', 'Bukayo Saka', 'Bukayo Saka', 'Bukayo Saka', 'Jude Bellingham'], 
    'minute': [3, 18, 37, 46, 87, 98], 
    'own_goal': [False, False, False, False, False, False], 
    'penalty': [False, False, False, False, True, False]
})

new_data_fr = pd.DataFrame({
    'date': pd.to_datetime('2026-07-18'), 
    'home_team': 'France',                
    'away_team': 'England',               
    'team': 'France',                    
    'scorer': ['Kylian Mbappé', 'Kylian Mbappé', 'Bradley Barcola', 'Ousmane Dembélé'], 
    'minute': [48, 66, 54, 96],
    'own_goal': [False, False, False, False], 
    'penalty': [False, False, False, False]
})


wc_data = pd.DataFrame({
    'date': pd.to_datetime('2026-07-19'), 
    'home_team': 'Spain',                
    'away_team': 'Argentina',               
    'team': 'Spain',                    
    'scorer': 'Ferran Torres', 
    'minute': 106, 
    'own_goal': [False], 
    'penalty': [False]
})

df = pd.concat([df, new_data_eng], ignore_index = True)
df = pd.concat([df, new_data_fr], ignore_index = True)
df = pd.concat([df, wc_data], ignore_index = True)
```
