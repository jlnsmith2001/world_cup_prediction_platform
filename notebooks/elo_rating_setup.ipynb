```python
elo_name_map = {
    "Curaçao": "Curacao",
    "Czech Republic": "Czechia",
    "Republic of Ireland": "Ireland",
    "Timor-Leste": "East_Timor",
    'São Tomé and Príncipe': "Sao_Tome_and_Principe",
    "United States Virgin Islands": "US_Virgin_Islands"
}
```

```python
countries = sorted(
    set(cleaned_df["home_team"]) |
    set(cleaned_df["away_team"])
)

all_matches = []
failed_countries = []

for country in countries:
    try:
        elo_country = elo_name_map.get(country, country) 
        elo_country = elo_country.replace(" ", "_")

        df = eloratings.country_matches_data(elo_country)
        all_matches.append(df)
        print(f"yep {country}")

    except Exception:
        print(f"nope {country}")
        failed_countries.append(country)

```
